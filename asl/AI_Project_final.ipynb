```python
# model = Sequential()
# model.add(Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=(64, 64,3)))
# model.add(Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
# model.add(MaxPool2D(pool_size=2, strides=2, padding='same'))
# model.add(Conv2D(filters = 64, kernel_size=3, padding='same', activation='relu'))
# model.add(MaxPool2D(pool_size=2, strides=2, padding='same'))
# model.add(Conv2D(filters = 128, kernel_size=3, padding='same', activation='relu'))
# model.add(MaxPool2D(pool_size=2, strides=2 ,padding='same'))
# #model.add(Flatten())
# #model.add(BatchNormalization())
# ############
# model.add(Flatten())
# ###
# model.add(Dense(units=526, activation='relu'))
# model.add(Dropout(0.5))
# #model.add(BatchNormalization())
# model.add(Dense(units=128, activation='relu'))
# model.add(Dropout(0.5))
# #model.add(BatchNormalization())
# model.add(Dense(units=29, activation='relu'))
# model.add(Dropout(0.5))
# #model.add(BatchNormalization())
# model.add(Dense(units = 29, activation='softmax'))

# model.summary()





#####################################################################################
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.models import Model

input_shape = (64, 64, 3)
num_classes = 29

inputs = Input(shape=input_shape)
x = Conv2D(filters=12, kernel_size=3, padding='same', activation='relu', name='conv1')(inputs)
x = Conv2D(filters=9, kernel_size=3, padding='same', activation='relu', name='conv2')(x)
x = MaxPooling2D(pool_size=2, strides=2, padding='same', name='pool1')(x)
x = Conv2D(filters=11, kernel_size=3, padding='same', activation='relu', name='conv3')(x)
x = MaxPooling2D(pool_size=2, strides=2, padding='same', name='pool2')(x)
x = Conv2D(filters=7, kernel_size=3, padding='same', activation='relu', name='conv4')(x)
x = MaxPooling2D(pool_size=2, strides=2, padding='same', name='pool3')(x)

x = Flatten()(x)
x = Dense(units=2, activation='relu', name='dense1')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
x = Dense(units=4, activation='relu', name='dense2')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
x = Dense(units=13, activation='relu', name='dense3')(x)
x = Dropout(0.5)(x)
x = BatchNormalization()(x)
outputs = Dense(units=num_classes, activation='softmax', name='output')(x)

model = Model(inputs=inputs, outputs=outputs)
model.summary()


#########################################################################
# from tensorflow.keras.models import Sequential
# from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
# from qkeras import QDense, quantized_bits, quantized_relu, QActivation
# input_shape = (64, 64, 3)
# num_classes = 29

# model = Sequential()

# model.add(Conv2D(filters=20, kernel_size=3, padding='same', activation='relu', name='conv1', input_shape=input_shape))
# model.add(Conv2D(filters=12, kernel_size=3, padding='same', activation='relu', name='conv2'))
# model.add(MaxPooling2D(pool_size=2, strides=2, padding='same', name='pool1'))
# model.add(Conv2D(filters=22, kernel_size=3, padding='same', activation='relu', name='conv3'))
# model.add(MaxPooling2D(pool_size=2, strides=2, padding='same', name='pool2'))
# model.add(Conv2D(filters=18, kernel_size=3, padding='same', activation='relu', name='conv4'))
# model.add(MaxPooling2D(pool_size=2, strides=2, padding='same', name='pool3'))

# model.add(Flatten())
# model.add(Dense(units=2, activation='relu', name='dense1'))
# model.add(Dropout(0.5))
# model.add(Dense(units=40, activation='relu', name='dense2'))
# model.add(Dropout(0.5))
# model.add(Dense(units=29, activation='relu', name='dense3'))
# model.add(Dropout(0.5))
# model.add(Dense(units=num_classes, activation='softmax', name='output'))

# model.summary()
```

```python
for layer in model.layers:
    if layer.__class__.__name__ in ['Conv2D', 'Dense']:
        w = layer.get_weights()[0]
        layersize = np.prod(w.shape)
        print("{}: {}".format(layer.name, layersize))  # 0 = weights, 1 = biases
        if layersize > 4096:  # assuming that shape[0] is batch, i.e., 'None'
            print("Layer {} is too large ({}), are you sure you want to train?".format(layer.name, layersize))
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale = 1./255, validation_split = .2)

valid_generator = datagen.flow_from_directory(
    'asl_alphabet_train/asl_alphabet_train',
    subset="validation",
    target_size = (64, 64),
    shuffle="true"
)

train_generator = datagen.flow_from_directory(
    'asl_alphabet_train/asl_alphabet_train',
    subset="training",
    target_size = (64, 64),
    shuffle="true"
)
```

```python
from tensorflow.keras.optimizers import Adadelta

model.compile(
    optimizer=Adadelta(learning_rate= 0.1),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history=model.fit(
    train_generator,
    epochs = 1,
    verbose=1,
    validation_data=valid_generator
)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
from glob import glob
import numpy as np
import random
import cv2

TRAIN_DIR = 'asl_alphabet_train/asl_alphabet_train'

CLASSES = [folder[len(TRAIN_DIR) + 1:] for folder in glob(TRAIN_DIR + '/*')]
CLASSES.sort()

cols = 5
rows = int(np.ceil(len(CLASSES) / cols))
fig = plt.figure(figsize = (16, 20))

for i in range(len(CLASSES)):
  cls = CLASSES[i];
  img_path = TRAIN_DIR + '/' + cls + '/**'
  path_contents = glob(img_path)

  imgs = random.sample(path_contents, 1)
  sp  =  plt.subplot(rows, cols, i+1)
  plt.imshow(cv2.imread(imgs[0]))
  plt.title(cls)
  sp.axis('off')
plt.show()
```
